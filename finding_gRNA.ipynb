```python
# count sequence function for all files in the directory
import os
import gzip

def count_sequences_in_fastq(file_path, sequences_to_find):

    sequence_counts = {sequence: 0 for sequence in sequences_to_find}

    with gzip.open(file_path, 'rt') as file:
        for i, line in enumerate(file):
            if i % 4 == 1:  # Sequence line
                read_sequence = line.strip()
                for sequence in sequence_counts.keys():
                    if sequence in read_sequence:
                        sequence_counts[sequence] += 1

    return sequence_counts


def count_sequences_in_directory(directory, sequences_to_find):
    sequence_counts = {sequence: 0 for sequence in sequences_to_find}

    for file_name in os.listdir(directory):
        file_path = os.path.join(directory, file_name)
        file_sequence_counts = count_sequences_in_fastq(file_path, sequences_to_find)

        for sequence in sequence_counts.keys():
            sequence_counts[sequence] += file_sequence_counts[sequence]

    return sequence_counts


count_sequences_in_directory('/wrk/jacky/fastq/origin', sequences_to_check)

```
